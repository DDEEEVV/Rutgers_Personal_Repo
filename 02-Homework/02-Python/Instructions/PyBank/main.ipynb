```python
"""
Davit Cuedari 
Rutgers ~ FinTech BootCamp --- Python HomeWork --- due Saturday, March 26, 2022

The task is to create a Python script that analyzes the records to calculate each of the following:

        1- The total number of months included in the dataset.

        2- The net total amount of Profit/Losses over the entire period.

        3- The average of the changes in Profit/Losses over the entire period.

        4- The greatest increase in profits (date and amount) over the entire period.

        5- The greatest decrease in losses (date and amount) over the entire period.

And have these results in an output file.txt: 
  
Financial Analysis
----------------------------
Total Months: 86
Total: $38382578
Average  Change: $-2315.12
Greatest Increase in Profits: Feb-2012 ($1926159)
Greatest Decrease in Profits: Sep-2013 ($-2196167)
"""
# To start, first it is necessary to import a few modules: ps: maybe I am importing some extra just in case haha!

import os
import csv
from numpy import average
import numpy_financial as npf

# After importing the module we have to open the file with csv function 
csvpath = os.path.join("/Users/ddevii/Rutgers/DC_Python_HomeWork/PyBank/Resources/budget_data.csv")

# I have to set an output file where I want the results of the exercise to be written:
output_document_path = "/Users/ddevii/Rutgers/DC_Python_HomeWork/PyBank/Resources/output_document.txt"

# I need to create a few variables that I am going to transfer and fill with data later on the exercise:
count_months = []
profit_loss_values = 0
net_total_pnl = 0
changes_in_pnl = []
changes_in_months_list = []
last_pnl = 0
last_pnl_list = []
average = 0
biggest_loss = ["", 10000000000000]
biggest_profit = ["", 0]

#opening the csv file and extracting data from it to use to find what is needed:
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

#to avoid counting the header with the total use:
    header = next(csvreader)
    for row in csvreader:

#to count the months we need to add them in a new list only with months and the length of that list is the number of months:
        count_months.append(row[0])
        #to count the sum of all the amounts of Profit/Losses each month
        profit_loss_values = int(row[1])
        net_total_pnl += profit_loss_values

#to calculate the average of the changes in Profit/Losses
        changes_in_pnl = int(row[1]) - last_pnl
        last_pnl = int(row[1])
        last_pnl_list = last_pnl_list + [changes_in_pnl]
        changes_in_months_list = [changes_in_months_list] +[row[0]]
        
        
        
        
#to find the greatest increase in profits (date and amount):
      
        if changes_in_pnl > biggest_profit[1]:
                biggest_profit[0] = row[0]
                biggest_profit[1] = changes_in_pnl

#to find the greatest decrease in profits (date and amount):

        if changes_in_pnl < biggest_loss[1]:
                biggest_loss[0] = row[0]
                biggest_loss[1] = changes_in_pnl

average =  (sum(last_pnl_list) - (last_pnl_list[0])) / len(count_months[1:])

#Printing the results into a text file:
with open(output_document_path, "w") as txt_file:
        txt_file.write(f"Financial Analysis\n")
        txt_file.write(f"------------------------------------------\n")
        txt_file.write(f"Total months: {len(count_months)}\n")
        txt_file.write(f"Total: ${net_total_pnl :,}\n")
        txt_file.write(f"Average Change: ${round(average, 2)}\n")
        txt_file.write(f"Biggest Increase in Profits: {biggest_profit[0]} $({biggest_profit[1]})\n")
        txt_file.write(f"Greatest Decrease in Profits: {biggest_loss[0]} $({biggest_loss[1]})\n")



print(f"The total number of months in the dataset is: {len(count_months)}")
print(f"The net total amount of Profit/Losses is: ${net_total_pnl :,}")
print(f"The average of the changes in Profit/Losses is: ${round(average, 2)}")
print(f"The biggest increase in profits is: {biggest_profit[0]} $({biggest_profit[1]})")
print(f"The greatest decrease in profits is: {biggest_loss[0]} $({biggest_loss[1]})")
```
