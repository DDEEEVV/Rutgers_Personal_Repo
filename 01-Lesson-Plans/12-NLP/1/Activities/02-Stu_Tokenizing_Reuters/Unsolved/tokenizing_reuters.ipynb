# Student Activity: Tokenizing Reuters

```python
from nltk.corpus import reuters
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd

# Code to download corpora
import nltk
nltk.download('reuters')
nltk.download('punkt')
```

```python
# Search through all categories
print(reuters.categories())
```

```python
# Get all fileids associated with cpi
cpi_id = reuters.fileids(categories = "cpi")
cpi_id
```

```python
raw_stories = [reuters.raw(id) for id in cpi_id]
len(raw_stories)
```

```python
sentence_tokenized = [sent_tokenize(story) for story in raw_stories ]

words_tokenized = [ [word_tokenize(word) for word in sentence]for sentence in sentence_tokenized]
print(words_tokenized)
```
