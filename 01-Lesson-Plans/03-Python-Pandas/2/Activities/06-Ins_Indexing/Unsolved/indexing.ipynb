### Instructor Demo: Indexing

This program reads performs several slice and dice operations through indexing via loc and iloc functions.

```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
```

### Read CSV in as DataFrame

```python
# Set the file path
csvpath = Path("../Resources/people.csv")

# Read in the CSV as a DataFrame
people_csvdata = pd.read_csv(csvpath)
```

```python
people_csvdata.head(2)
```

```python
people_csvdata.describe()
```

```python
people_csvdata.describe(include="all")
```

```python
#selecting the first row of the DataFrame with .iloc stands for intiger location
people_csvdata.iloc[0]
```

```python
# and in case we like to have more rows: with brackets just like the list. most important and most used is loc for locate used for intigers as well
people_csvdata.iloc[0:3, [0,2,6]]
```

```python
people_csvdata.iloc[0:10]
```

```python
people_csvdata.iloc[-1]
```

```python
#selecting the first column of the DataFrame:
people_csvdata.iloc[:,0].head()
```

```python
#selecting the second column of the DataFrame:
people_csvdata.iloc[:,1].head()
```

```python
#selecting the last column of the DataFrame:
people_csvdata.iloc[:,-1].head()
```

```python
#selecting the first 2-columns of the DataFrame:
#iloc [row_start : row_end, columns_start, columns_end or strings ([1, 2, 3....])]
people_csvdata.iloc[:,0:2].head()
```

```python
# Select the 1st, 5th, 8th, 22nd rows of the 1st 4th and 6th columns.
people_csvdata.iloc[[0, 4, 7, 21], [0, 3, 5]]
```

```python
# Select the first 5 rows of the 3rd, 4th, and 5th columns of the DataFrame
people_csvdata.iloc[0:4, [2,3,4]]
```
